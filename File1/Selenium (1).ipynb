Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and 
salary filter. 
You have to scrape data for “Data Scientist” designation for first 10 job results. 
You have to scrape the job-title, job-location, company name, experience required. 
The location filter to be used is “Delhi/NCR”. The salary filter to be used is “3-6” lakhs 
The task will be done as shown in the below steps: 
1. first get the web page https://www.naukri.com/
2. Enter “Data Scientist” in “Skill, Designations, and Companies” field. 
3. Then click the search button. 
4. Then apply the location filter and salary filter by checking the respective boxes 
5. Then scrape the data for the first 10 jobs results you get. 
6. Finally create a dataframe of the scraped data.

```python
driver=webdriver.Chrome()
```

```python
# opening the naukri page on automated chrome browser
driver.get("https://www.naukri.com/")
```

```python
# entering designation as required in the question
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys("Data Scientist")
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
# Applying location filter
location_filter=driver.find_element(By.XPATH,'/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[3]/label/i')
location_filter.click()
```

```python
# Applying Salary filter
salary_filter=driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[6]/div[2]/div[2]/label/i")
salary_filter.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
# scraping job title from the given page
title_tags=driver.find_elements(By.XPATH,'//a[@class="title "]')[:10]
for i in title_tags:
    title=i.text
    job_title.append(title)
    
# scraping job location from the given page
loc_tags=driver.find_elements(By.XPATH,'//span[@class="locWdth"]')[:10]
for i in loc_tags:
    location=i.text
    job_location.append(location)

# scraping company name from the given page
comp=driver.find_elements(By.XPATH,'//a[@class=" comp-name mw-25"]')[:10]
for i in comp:
    company=i.text
    company_name.append(company)
    
# scraping experience required from the given page
exp=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')[:10]
for i in exp:
    experience=i.text
    experience_required.append(experience)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
df=pd.DataFrame({'Title':job_title,'Location':job_location,'Company':company_name,'Experience':experience_required})
df
```

field.

3. Price
As shown in the below image, you have to scrape the above attributes

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com/")
```

```python
import time
```

```python
# searching for sneakers
sneak=driver.find_element(By.CLASS_NAME,"_3704LK")
sneak.send_keys("Sneakers")
```

```python
search=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search.click()
```

```python
brand=[]
start=0
end=3
for page in range(start,end):
    b=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in b:
        brand.append(i.text)
    nextb=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    nextb.click()
    time.sleep(3)
    
product_desc=[]
start=0
end=3
for page in range(start,end):
    pd=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in pd:
        product_desc.append(i.text)
    nextb=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    nextb.click()
    time.sleep(3)
    
price=[]
start=0
end=3
for page in range(start,end):
    pr=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in pr:
        price.append(i.text)
    nextb=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    nextb.click()
    time.sleep(3)
```

```python
len(brand),len(product_desc),len(price)
```

```python
# Find the maximum length
max_len = max(len(brand), len(product_desc), len(price))

# Pad the shorter lists with null values
brand = brand + [None] * (max_len - len(brand))
product_desc = product_desc + [None] * (max_len - len(product_desc))
price = price + [None] * (max_len - len(price))

# Create the DataFrame
df1 = pd.DataFrame({'Brand': brand, 'Product Description': product_desc, 'Price': price})
print(df1)
```

```python
# scraping quotes
quotes=[]
start=0
end=9
for page in range(start,end):
    q=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in q:
        quotes.append(i.text)
    nextt=driver.find_element(By.XPATH,'//li[@class="next"]')
    nextt.click()
    time.sleep(3)
    
```

```python
len(quotes)
```

```python
# scraping authors
author=[]
start=0
end=9
for page in range(start,end):
    a=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in a:
        author.append(i.text)
    nextt=driver.find_element(By.XPATH,'//li[@class="next"]')
    nextt.click()
    time.sleep(3)
    
```

```python
len(author)
```

```python
# scraping type of quotes
typesq=[]
start=0
end=9
for page in range(start,end):
    t=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in t:
        typesq.append(i.text)
    nextt=driver.find_element(By.XPATH,'//li[@class="next"]')
    nextt.click()
    time.sleep(3)
```
